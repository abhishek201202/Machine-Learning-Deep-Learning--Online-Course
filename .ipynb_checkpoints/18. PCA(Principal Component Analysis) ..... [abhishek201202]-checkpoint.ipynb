# 1) 2-D   to   2-D

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x1 = np.array([1 , 2 , 3 , 4 , 5 , 6])
x2 = np.array([7.5 , 11 , 16 , 18 , 20 , 26])
plt.scatter(x1 , x2)
plt.show()
```

```python
x = np.c_[x1 , x2]
x
```

```python
pca = PCA(n_components = 2)
x_transformed = pca.fit_transform(x)
x_transformed
```

```python
pca.components_
```

```python
## to get vack data into the original dimensions
pca.inverse_transform(x_transformed)
```

# 3) 3-D

### 3a) plot 3-D features

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243)
## random data that we are generating is same irrespective of 
## how many times you run
```

```python
mean_vec1 = np.array([0 , 0 , 0])
cov_mat1 = np.array([[1 , 0 , 0] , [0 , 1 , 0] , [0 , 0 , 1]])
## covariance matrix
class1 = np.random.multivariate_normal(mean_vec1 , cov_mat1 , 100)
```

```python
mean_vec2 = np.array([1 , 1 , 1])
cov_mat2 = np.array([[1 , 0 , 0] , [0 , 1 , 0] , [0 , 0 , 1]])
class2 = np.random.multivariate_normal(mean_vec2 , cov_mat2, 100)
class2.shape
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D , proj3d

fig = plt.figure(figsize = (8 , 8))
## i need a 8 by 8 inches figure
ax = fig.add_subplot(111 , projection = '3d')
## divide your figure into 1 row and 1 col and plot the figure into the 1st position row wise
## this line say that i want to add subplot to above figure 
## and 111 means that 1 row , 1 col and 1st position i try to fill
## ex : 234 means 
## first divide ( 8 x 8 ) ... figure into 2 rows and 3 columns and pick the 4th position cell
## and in this cell we will plot our figure 
ax.plot(class1[ : , 0] , class1[ : , 1] , class1[ : , 2] , 'o')
ax.plot(class2[: , 0] , class2[ : , 1] , class2[ : , 2] , '^')
plt.show()
```

### 3b) transform to new features

```python
all_data = np.concatenate((class1 , class2))
all_data
all_data.shape
```

```python
pca = PCA(n_components = 2)
transformed_data = pca.fit_transform(all_data)
transformed_data
```

```python
pca.components_
```

```python
plt.plot(transformed_data[0 : 100, 0] , transformed_data[0:100 , 1] , "o")
plt.plot(transformed_data[100 : 200, 0] , transformed_data[100:200 , 1] , '^')
plt.show()
```

### 3c) invert back to original features

```python
x_approx = pca.inverse_transform(transformed_data)
fig = plt.figure(figsize = (8 , 8))
ax = fig.add_subplot(111 , projection = '3d')
ax.plot(x_approx[ : , 0] , x_approx[ : , 1] , x_approx[ : , 2] , 'o')
plt.show()
## this all data point now lies on a plane
## while inverting back from 3-D to 2-D
## in case when we plot it from 2-D to 1-D ..... it was liying on the single line
```

```python
## equation of the above plane is 
## a*x + b*y + c*z = 1
a = -0.409689
b = 7.2827
c = - 7.1008
i = 20
a * x_approx[i][0] + b * x_approx[i][1] + c * x_approx[i][2]
```

# 4) PCA on Breast Cancer datasets

```python
from sklearn import decomposition , ensemble , datasets , linear_model
import numpy as np
import time
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
breast_cancer = datasets.load_breast_cancer()
```

```python
x = breast_cancer.data
x.shape
```

```python
sc = StandardScaler()
x_std = sc.fit_transform(x)
x_std.shape
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_std , breast_cancer.target , random_state = 0)
```

```python
pca = decomposition.PCA(n_components = 15)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
lr = linear_model.LogisticRegression()
start = time.time()
lr.fit(x_train , y_train)
end = time.time()
## this time is depends upon what all other programs are running on your processor
print("time taken iss : " , end - start)
lr.score(x_test , y_test)
```

```python
lr = linear_model.LogisticRegression()
start = time.time()
lr.fit(x_train_pca , y_train)
end = time.time()
print("time taken iss : " , end - start)
lr.score(x_test_pca , y_test)
```

```python
pca.explained_variance_
```

```python
all_data_transpose = all_data.T
cov = np.cov(all_data_transpose)
cov.shape
```

```python
cov
```

```python
eig_val , eig_vectors = np.linalg.eig(cov)
## eigen vector are column wise 
eig_val , eig_vectors
```

```python
eig_val_vector_pair = []
for i in range(len(eig_val)):
    eig_vec = eig_vectors[ : , i]
    eig_val_vector_pair.append((eig_val[i] , eig_vec))
eig_val_vector_pair.sort(reverse = True)
eig_val_vector_pair
```
