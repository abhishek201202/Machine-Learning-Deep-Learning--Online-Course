## 4) Data manipulation on rows

### a) drop with position no.

```python
## remove the row which has label something
temp = df.drop(1) ## bydefault it return a copy after deleting 1st row
df.drop(3 , inplace = True)
print(temp)
df.head()
```

### b) drop with index/label no.

```python
print(df.index)
df.index[0] , df.index[3]
```

```python
df.drop(df.index[0] , inplace = True)
```

```python
df.drop(df.index[[0,1,3]],inplace = True)  ## delete all row with index 0 , 1 , 3
df.head()
```

### c) drop and display with some condition

```python
## df[boolean array] apply on row
print(df.sl > 5)
print(df[df.sl > 5])  ## give only those for which the condition is true
df[df.flower_type == 'Iris-setosa'].describe()
```

### d) iloc VS loc

```python
print(df.head())
print()
# to get the top row
print(df.iloc[0])  ## work on the basis of position
print()
## print(df.loc[0])
print(df.loc[35])  ## work on the basis of label
```

## 5) Data manipulation on columns

```python
## drop a column 
## axis = 1 means here that look at column wise 
## if we don't pass the axis then it will try to find in row
## which will give error
df.drop('flower_type',axis = 1 ,inplace = True)
df
```

## 6) NAN handling

### a) delete some rows

```python
df.iloc[2:4 , 1:3] = np.nan
df.head()
```

```python
df.describe()
```

```python
## simply drop row which has NAN
df.dropna(inplace = True)
df.head()
```

```python
df.reset_index(drop = True , inplace = True)
df.head()
```

### b) fill some entry on that place

```python
df.iloc[2:4 , 1:5] = np.nan
df.head()
```

```python
## fill the entry in place of NAN
# df.fillna(-1,inplace = True) ## replace all na with -1
df.pw.fillna(df.pw.mean(),inplace = True)
df.pl.fillna(df.pl.mean(),inplace = True)
df.head()
# df.describe()
```

## 7) Handling strings data

```python
df["gender"] = "Female"
df.iloc[1:3 , 5] = "Male"
df.head()
```

# 8) Apply some function on some column

```python
## apply some function on some column
def f(s):
    if s == "Male":
        return 0
    else :
        return 1
df["Sex"] = df.gender.apply(f)
del df["gender"]
df.head()
```
